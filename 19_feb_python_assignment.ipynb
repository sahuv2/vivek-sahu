### Q3. What is App routing in Flask? Why do we use app routes? 


n Flask, app routing refers to the process of mapping URL endpoints to specific functions or views within a Flask application.

This is achieved by using decorators in Flask, specifically the @app.route decorator, which is used to define routes.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/contact')
def contact():
    return 'Contact us at example@example.com'

if __name__ == '__main__':
    app.run()
```
